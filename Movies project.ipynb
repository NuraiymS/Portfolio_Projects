```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  #Adjust the configuration of the plots we'll create
```

```python
df = pd.read_csv(r'/Users/nuraiymsardarbekova/Downloads/movies.csv')

df.head()
```

```python
# checking to see if there is any missing data


for col in df.columns:
    missing_val = np.mean(df[col].isnull())
    f_missing_val = f"{missing_val:.2%}"
    print('{} - {}%'.format(col, f_missing_val))
    
```

```python
df.dtypes
```

```python
#Drop Nan cells to change type of columns 

df = df.dropna(subset = 'budget')

df = df.dropna(subset = 'gross')

df = df.dropna(subset = 'votes')
```

```python
#change data type of columns

df['votes'] = df['votes'].astype('int64')

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#create a correct year column by changing the type of the 'released' column to str, then split it and extract 2nd value

df['released'] = df['released'].astype(str).str.split().str[2]
```

```python
df.head(10)
```

```python

df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
df.duplicated().head()
```

```python
# Budget high correlation
# Company high correlation


#scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross earnings')

plt.xlabel('Budget for film')

plt.ylabel('Gross earnings')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df)

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
#looking at correlation

corr_matrix = df.corr()

corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation Matric')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# try to find different correlations as well, change objects to numbers

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
# Now check correlation again

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Organized the correlation

corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

#corr_pairs
```

```python
# Let's sort the values

sorted_pairs =  corr_pairs.sort_values()

#sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
