# Forbes billionaires 2022

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/nuraiymsardarbekova/Downloads/2022_forbes_billionaires.csv')

df
```

```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
#remove the $ signs and convert the networth column to float:

df["networth"] = df["networth"].str.strip("$")
df["networth"] = df["networth"].str.strip("B")
df["networth"] = df["networth"].astype(float)
```

```python
df.head()
```

```python
# Top 10

df1 = df.sort_values(by = ["networth"], ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(df1["name"], df1["networth"])
plt.title("Net Worth by Name")
plt.xlabel("Name")
plt.ylabel("Net Worth")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#top 5 sectors with the most number of billionaires

a = df["source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "grey", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors, autopct='%1.0f%%')  # Added autopct parameter
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Sectors to Become a Billionaire", fontsize=20)
plt.show()
```

```python
b = df["industry"].value_counts().head()
index = b.index
industries = b.values
custom_colors = ["skyblue", "green", 'tomato', "pink", "red"]
plt.figure(figsize=(5, 5))
plt.pie(industries, labels=index, colors=custom_colors, autopct=lambda p: '{:.0f}'.format(p * sum(industries) / 100))
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with Most Number of Billionaires", fontsize=20)
plt.show()
```

```python
c = df["country"].value_counts().head()
index = c.index
countries = c.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "pink", "red"]
plt.figure(figsize=(5, 5))
plt.pie(countries, labels=index, colors=custom_colors, autopct=lambda p: '{:.0f}'.format(p * sum(countries) / 100))
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```
